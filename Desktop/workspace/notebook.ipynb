```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pib_global =  pd.read_csv('gapminder_all.csv')
pib_global.head()
```

Comparamos el PIB de Ecuador y Chile 

```python
pib_ecuador = pib_global[(pib_global.country == 'Ecuador')|(pib_global.country == 'Chile')]
pib_ecuador
                         
```

```python
pib_ecuador.columns[pib_ecuador.columns.str.startswith('gdp')]
```

Para obtener el mismo resultado:

```python
col_int=[]
for col in pib_ecuador.columns.values: #values me da como un array,columnas llamadas col
    if col.startswith('gdp'):# haciendo este buckle con array ya se hace str
        col_int.append(col)# en la col vac√≠a llena 
```

Ahora lo mismo le hago lista:

```python
mi_lista = [col for col in pib_ecuador.columns.values if col.startswith('gdp')]
mi_lista
```

```python
nombres_pais = pd.DataFrame(pib_ecuador.country)
nombres_pais
```

```python
pib_ecuador = pib_ecuador[mi_lista].T
pib_ecuador #61 es el √≠ndice de la columna 
```

```python
pib_ecuador.columns = nombres_pais.country
pib_ecuador
```

```python
pib_ecuador.index
```

```python
pib_ecuador.reset_index(inplace = True)
```

```python
pib_ecuador.rename(columns = {'index':'anio'}, inplace = True) #inplace = True para que gruarde la info en el mismo DataFrame pib_ecuador 
pib_ecuador
```

```python
pib_ecuador['anio']= pib_ecuador['anio'].str.replace('gdpPercap_','')
pib_ecuador #Ya me sale solo el a√±o 
```

```python
pib_ecuador['anio'] = pib_ecuador.anio.astype(int) 

# convirtiendo el a√±o de str a entero 
```

```python
pib_ecuador.dtypes
```

```python
pib_ecuador.plot(x = 'anio', y= 'Chile')
plt.xlabel('A√±o')#asume las etiquetas solo ara el ultimo gr√°fico
plt.ylabel('PIB per c√°pita')
pib_ecuador.plot(x = 'anio', y= 'Ecuador')
plt.xlabel('A√±o')#asume las etiquetas solo ara el ultimo gr√°fico
plt.ylabel('PIB per c√°pita')
```

```python
fig = plt.figure()#un gr√°fico dentro de la figura 
eje = fig.add_subplot(1,1,1) #n filas #n columnas #indice 46:26
pib_ecuador.plot(x = 'anio', y= 'Chile',ax = eje, linestyle = '-.')
pib_ecuador.plot(x = 'anio', y= 'Ecuador',ax = eje, linestyle = '--')
plt.xlabel('A√±o')#asume las etiquetas solo ara el ultimo gr√°fico
plt.ylabel('PIB per c√°pita')
```

```python
fig= plt.figure()
eje = fig.add_subplot(1,2,1)
pib_ecuador.plot(x= 'anio', y = 'Chile', ax= eje, ls = '-.')

eje2 = fig.add_subplot(1,2,2,sharey = eje)
pib_ecuador.plot(x = 'anio', y = 'Ecuador', ax = eje2, ls = 'dotted')
eje.set_xlabel('A√±o')
eje2.set_xlabel('A√±o')

# mas opciones en matplotlib
```
